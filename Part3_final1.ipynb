###plot_all_histograms() 
plot_all_histograms() accepts a pandas DataFrame and plots a histogram for the first 100 column on a 10 x 10 grid (if there are less than 100 columns, then it just plots all the columns). 

This is useful for quickly evaluating all of the features in my database. If there is something wrong with a column's data, so that it can't make a histogram (e.g., the data is not numeric) then that feature is skipped and it prints a warning message. 

```python
def plot_all_histograms(df, show_names=False):
    #this will plot a 10 x 10 subplot with a histogram of each variable in each subplot
    for x in range(100):
        if x < df.shape[1]:
            plt.subplot(10,10,x+1)
            try: 
                plt.hist(df[df.columns[x]]) # if it exists
            except:
                print "Didn't plot hist for {0}".format(df.columns[x])
            # format the plots so that they aren't cluttered. 
            remove_border()
            ax = plt.gca()
            ax.spines['top'].set_visible(True)
            ax.spines['right'].set_visible(True)
            ax.spines['left'].set_visible(True)
            ax.spines['bottom'].set_visible(True)
            if show_names:
                plt.title(df.columns[x], size = 8) #plot name of feature
        else:
            return
    if show_names:
        plt.tight_layout()
```

### Factor Analysis and Varimax Rotation: a good method for dimensionality reduction and interpreting features

Since we know that our feature space has lots of correlations, we can actualy represent those data in a lower dimension space while still explaining all of the variance of the data. Methods like Principal Components Analysis (PCA) and Factor Analysis (FA) are good for this. These methods leverage covariances between variables and transform your original data into new "factors" or "components". Each factor is orthogonal to the others, and is represented as a linear combination of your original variables. In other words, you capture all the shared variance across variables into single variables. 

As an example, I can do PCA on all of the features in ACS data set. Plotted below you can see that although there were originally 102 different observations with over 300 features on each, about 90% fo the variance in our observations is explained with just about 20 components ("factors"). Then there is a "long tail" which requires another 30 or 40 components to explain the remaining 10% of variance. 

```python
# do PCA on full feature space
from sklearn.decomposition import PCA
fullpca = PCA()
fullpca.fit(zscore(acs_est_df.values))
```

```python
def visualize_pca(pca):
    # make scree plot and cumulative variance plot on a pca matrix
    plt.figure()
    plt.subplot(1,2,1)
    plt.plot(pca.explained_variance_ratio_)#cumulative variance
    plt.title('Scree Plot for PCA on All Features')
    plt.xlabel('Component Number')
    plt.ylabel('Variance Explained')
    plt.xlim([0,pca.explained_variance_ratio_.shape[0]])
    plt.subplot(1,2,2)
    plt.title('More Variance Explained with additional PCA Components')
    plt.xlabel('Component Number')
    plt.ylabel('Cumulative Variance Explained')
    plt.plot(np.cumsum(pca.explained_variance_ratio_))#cumulative variance
    plt.ylim([0,1])
    plt.xlim([0,pca.explained_variance_ratio_.shape[0]])
    plt.tight_layout()
    
```

###Varimax rotation to the rescue!
A nifty technique that we use at Prophecy Sciences to help interepret our dimensionality reduction is the varimax rotation.  http://en.wikipedia.org/wiki/Varimax_rotation. Varimax is an orthogonal rotation of your factor axes (or principal component axes) that forces variables to have either high or low loadings with each factor (it accomplishes this by maximizing the variance of the squared loadings within each factor). This is awesome and makes it easier to interpret your axes, because after varimax rotation each factor tends to be associated with a smaller number of high-loaded variables with small loadings given to all other variables.

This is how __Prophecy Sciences__ relates cognitive and personality differences in individuals to the hundreds of signals that we measure from playing our cognitive games. It helps us summarize our hundreds of measurements into about a dozen factors--which we call "Cognitive DNA" factors--each of which are easier to interpret. Using our neuroscience expertise, we are able to assign meaningful semantic labels like "Memory" and "Flexible Thinking" to these different factors by examining the common relationship between the high-loading variables in each factor.

###A Poor Man's Factor Analysis. 
After trying out some PCA and FA, and spending some time trying to interpret my factors without the benefit of varimax rotation, I decided that this was too greatly obscuring my ability to interpret the ACS data set. When using these data to model Lumosity scores, I wanted to error on the side of interpretability over predictive accuracy. 

Therefore, I wanted a way to group my variables into functionally-related clusters that were conceptually analagous to the factors one gets from Factor Analysis. I decided that a simple (non-optimal) way to do this is to simply group variables together which are highly correlated, as measured by pearson correlation coefficient. Maybe the groupings that I would find from this analysis would be easier to interpret than the factors from my FA analysis? 


```python
def get_correlated_feature_groups(df, criterion = 0.75, min_sublist_length = 2):
    """
    This is my poor man's factor analysis algorithm. 
    It produces a list of sublists of highly correlated features, and requires that each feature only is used once in one sublist. 
    This is the clunkiest code that I have written in for this case study, and has not been particularly optimized for speed or function. 
    It uses the following logic 1) does a for loop across all the features, 
                                2) identifies other features taht are correlated with that feature, 
                                3) checks that none of these features have already been assigned to a previous group
                                4) and appends those features as a new sublist 
    
    Inputs
        df, a dataframe. By default the columns of the dataframe are treated as separate 
                          variables and rows are different observations.
        criterion = this is the threshold for pearson correlation coefficient. 
                    
    Returns
        MasterList - a list of sublists, each sublist contains features which are 
                      correlated with eachother with a pearson correlation coefficient > criterion.
                      Every variable in a particular sublist is correlated with *at least one other variable* in its sublist
                      Every sublist has at least min_sublist_length number of features in it.
    """
    R= np.corrcoef(df.values, rowvar = 0) # this is the cross correlation matrix for all features 
    MasterList = []
    for feature in df.columns:
        if feature not in [item for sublist in MasterList for item in sublist]:
            k = df.columns.tolist().index(feature) # the numerical index for this feature in the R matrix
            correlations = (pd.DataFrame(R[:,k],index = df.columns)).copy() # for feature, a list of its correlations with all other features       
            correlated_features_idx = np.abs(correlations.values)>criterion # which features are correlated > criterion level
            new_features = df.columns.values[correlated_features_idx.flatten()] # pull these features into a list. 
            
            #this is a key step where I ensure that none of these new features have already been assigned to previous sublists
            new_features_ = [nf for nf in new_features if nf not in [item for sublist in MasterList for item in sublist] ]
            if len(new_features_)>=min_sublist_length:
                    MasterList.append(new_features_)
    return MasterList
```

```python
# FL is the 'factor list'. Each element of FL is a correlated sublist of features
FL = get_correlated_feature_groups(Zacs_est_df)
print "This analysis started with {0} features \nand found {1} major clusters of 2 or more features which are highly correlated with one another.".format(Zacs_est_df.shape[1],len(FL))
```

####Cross validated Accuracy Calculations
To evaluate the accuracy of my model I will use mean squared error (MSE) between my predictions of Lumosity scores and the true Lumosity scores for cities. 

The MSE is calculated via a __cross validation__ method. For every model, 60% of the observations were randomly chosen to model the data. That linear regression was then used to predict the remaining 40% of the data and a MSE was calcualted on those predictions. 

This is a rigorous way to evaluate the predictive ability of your model on future data, and helps avoid over-fitting to your particular set of observations. 

Here are some functions I will use to build and evaluate my linear regression models, but feel free to skip down to the awesome figure below. 

```python
def build_model(X,Y, number_of_features =None):
    """
        Build a multiple linear regression model on a subset of the data
            inputs: 
                X, Y -- are the indepenent and dependent variables we are modeling, they are np.arrays.
                number_of_features -- is how many columns of X we are including in the model. 
                                     If None, then we include ALL the variables given in X
            outputs:
                regr -- the LinearRegression sklearn object, fitted to xtrain and ytrain
                xtrain, ytrain -- the subsets of the data, randomly chosen, on which the model was fit
                xtest, ytest -- the subset of the data that were kept blind to the model. To be used for validation.
    """ 
    if number_of_features is not None:
        #number_of_features = X.shape[1]
        X = X[::, :number_of_features]
        
    # randomly choose a train and test set from the data
    xtrain, xtest, ytrain, ytest = train_test_split(X, Y, train_size=0.6)     
    
    #fit model              
    regr = linear_model.LinearRegression()
    regr.fit(xtrain,ytrain)
    return regr, xtrain, xtest, ytrain, ytest


def build_model_and_calculate_MSE(X, Y, number_of_features = 1, n_sim = 100):
    """
        This is a wrapper function around build_model() used to calcualte the change in model accuracy as you add additional features
        It uses a bootstrap approach and a cross-validation train/test approach to estimate the accuracy of of the model in terms of MSE
        
        Inputs
            X, Y, as in build_model()
            number_of_features -- as in build model, defaults to 1
            n_sim -- the number of simulations to be run. 
        Returns
            tuple (avearge MSE results, STD of MSE results)
                    These estimates come from running n_sim simulations on the data, each time taking a different random training set, 
                    and evaluating its accuracy on a a different random test set of the data. 
                    After running n_sim simulations we have an estimate of the distribution of the MSE of this particular model of the data. 
    """ 
    MSE_results =  np.array([np.nan]*n_sim)
    for sim in range(n_sim):
        X = X[::, :number_of_features]
        regr, xtrain, xtest, ytrain, ytest = build_model(X,Y, number_of_features =X.shape[1])
        y_predict = regr.predict(xtest)
        MSE_results[sim] = MSE(ytest, y_predict) # calculate the mean squared error for these predicitons
    return (round(np.mean(MSE_results),3), round(np.std(MSE_results),3))
```

```python
#prepare data
Y = data_aligned.Grand_Index.values # this is what we are hoping to predict
X = data_aligned[Zacs_select.columns].values

# Let's explore how changing the number of variables included in our models helps or hurts our predictions
num_features = X.shape[1]
MSE_list = [np.nan]*num_features
for k in range(num_features):
    MSE_list[k] = build_model_and_calculate_MSE(X,Y, number_of_features = k+1)

# plot MSE vs variable numbers
ax = plt.figure()
plt.errorbar(range(1,len(MSE_list)+1), zip(*MSE_list)[0], zip(*MSE_list)[1])
plt.title('Non-linear reduction in MSE as you add additional predictor variables')
ax = plt.gca()
ax.set_xlabel('Number of variables in the linear model')
ax.set_ylabel('Mean Squared Error of Lumosity Grand_Index (+- STD)')
```

Using these best features I made a final model. I wrote a few functions to help handle this and visualize the model:

1. prepareXandY() -- takes the ACS and SCS dataframes, aligns them with each other and returns X and Y matrices ready for modeling

2. visualize_predictions() --is all plotting. It plots scatter matrices of the model predictions and of the residuals. It also computes and shows MSE for the model. 

3. do_model() -- is a wrapper function. It calls prepareXandY(), build_model(), and visualize_predictions() so with one function you can complete all the steps and evaluate a particular model. 

 


```python
def prepareXandY(dfX, dfY, features_to_model = 3, targetY_label = 'Grand_Index'):
    if features_to_model is None:
        features_to_model = dfX.columns
    data_aligned = dfX.copy()
    data_aligned = data_aligned.join(dfY, how='inner') #align the msa labels
    data_aligned = data_aligned.sort()
    #prepare data
    Y = zscore(data_aligned[targetY_label].values) # this is what we are hoping to predict
    X = data_aligned[features_to_model].values
    print "X shape is : {0}".format(X.shape)
    print "Y shape is : {0}".format(Y.shape)
    return X, Y
```

```python
def do_model(X_data, Y_data, features_to_model = best_features, targetY_label = 'Grand_Index'): #(Zacs_est_df, scs_df, features_to_model = best_features)
    # prepare the data for learning
    X, Y = prepareXandY(Zacs_est_df, scs_df, features_to_model = best_features, targetY_label = 'Grand_Index')
    #build_model_and_calculate_MSE(X,Y, number_of_features = X.shape[1])
    
    # build the regression and visualize the results
    regr, xtrain, xtest, ytrain, ytest = build_model(X[::,:X.shape[1]], Y, number_of_features = X.shape[1])
    visualize_predictions(X[::,:X.shape[1]], Y, xtrain, ytrain, xtest, ytest, regr)
    return regr
```

```python
def visualize_predictions(X, Y, xtrain, ytrain, xtest, ytest, regr):
    """
        after building a model, send the results to this function to get a nice visualization of the predictions
    """
    #After building a model, 
        
    #evaluate predictive value on test set
    y_predict = regr.predict(xtest)
    
    fig = plt.figure(1)
    plt.title('Predicted Scores vs Actual Scores')
    plt.xlabel("Actual City Score")
    plt.ylabel('Predicted City Score')
    plt.scatter(ytest, y_predict, color ='r', alpha = 0.5)    
    plt.scatter(ytrain, regr.predict(xtrain), color= 'b', alpha = 0.3)
    remove_border(bottom = True, left = True)
    
    MSE_avg_std = build_model_and_calculate_MSE(X,Y, number_of_features = xtrain.shape[1])
    
    #quantify the quality of the prediction as R^2
    plt.text(ytest.min(),y_predict.max(), "R^2:  {0}".format(round(regr.score(xtest, ytest),3))) #
    #quantify the quality of the prediction as MSE
    plt.text(ytest.min(),y_predict.max()-1, "MSE:  {0} +- {1}".format(MSE_avg_std[0], MSE_avg_std[1]))
    plt.text(ytest.min(),y_predict.max()-2, "Number Features: {0}".format(xtrain.shape[1]))
    
    plt.figure(2)
    visualize_residuals(ytest,y_predict)

def visualize_residuals(ytest,y_predict):
    residuals = ytest - y_predict
    plt.title('Residuals as a function of Overall Scores')
    plt.xlabel("Actual City Score")
    plt.ylabel('Residual on prediction')
    plt.scatter(ytest, residuals)
    remove_border(bottom = True, left = True)
    r, p = scipy.stats.pearsonr(ytest,residuals)
    plt.text(ytest.min(),np.mean(residuals), "R^2:  {0}, p = {1}".format(round(r,3), round(p,4))) #
```

### Interepreting the Model

So now I have successfully predicted Lumosity scores using just 7 variables from the data set. 

Remember these 7 variables each are in the model representing a cluster of variables that are highly correlated and are linked by some meaningful factor. 



```python
def show_weights(best_features, regr, best_features_idx):
    weights = zip(best_features, regr.coef_, best_features_idx)
    weights = sorted(weights, key=lambda tup: np.abs(tup[1]), reverse= True)
    return weights #(feature_id, weight, ryan's factor)
```

```python
show_weights(best_features, regr, best_features_idx)
```
